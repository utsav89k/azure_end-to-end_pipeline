```python
df=spark.read.format("parquet").load("s3://pipeline-project-uk/bronze/FactStream")
```

```python
df.display()
```

```python
df.printSchema()
```

```python
from pyspark.sql.functions import col, when
```

```python
df = df.withColumn("is_carryable", when(col("device_type") == "Desktop", "No")\
    .when((col("device_type") == "Mobile") | (col("device_type") == "Smart Speaker"), "Yes")\
    .otherwise("No Data Found"))

```

```python
    # Arranging the Columns
    df = df.select(
        "stream_id",
        "user_id",
        "track_id",
        "date_key",
        "listen_duration",
        "device_type",
        "is_carryable",       
        "stream_timestamp"    
        )
```
